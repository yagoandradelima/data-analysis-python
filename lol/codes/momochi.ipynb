```python
# Reagrupando as colunas que vou utilizara efetivamente nos dados de análise

sup_data = full_data[["playername","league","split","teamname","position","champion","side","dpm","damagetakenperminute","damagemitigatedperminute","damageshare","damagetotowers","wardsplaced","wpm","wardskilled","wcpm","controlwardsbought","visionscore","vspm","totalgold","earnedgold","earned gpm","earnedgoldshare","goldspent","kills","deaths", "assists","goldat10","xpat10","csat10","golddiffat10","xpdiffat10","killsat10","assistsat10","deathsat10","goldat15","xpat15","csat15","golddiffat15","xpdiffat15","csdiffat15","killsat15","assistsat15","deathsat15","goldat20","xpat20","csat20","golddiffat20","xpdiffat20","csdiffat20","killsat20","assistsat20","deathsat20","goldat25","xpat25","csat25","golddiffat25","xpdiffat25","csdiffat25","killsat25","assistsat25","deathsat25"]]
```

```python
# Realizando o filtro final para o dataframe dos suportes do CD

sup_data = sup_data[(sup_data["league"] == "CD") & (sup_data["position"] == "sup")]
```

```python
# Criando dataframe de dados individuais do momochi

momochi_data = sup_data[sup_data["playername"] == "Momochi"]
```

```python
# Bloco único para criação do dataframe de KDA dos suportes do Circuito Desafiante

# Criação do df
kda_comparision = pd.DataFrame(columns=["playername","kda_total"])

# Criaçar das listas que receberão os dados do loop for
playername = []
kda_total = []
kda_medio = []
partidas = []

# loop for que vai percorrer cada nome de jogador e calcular seu KDA
for player in sup_data["playername"].unique():
    player_data = sup_data[sup_data["playername"] == player]

    # Criação do agregador de K, D e A dentro do campeonado
    player_kills = player_data.agg({"kills":"sum","deaths":"sum","assists":"sum"}).iloc[0]
    player_deaths = player_data.agg({"kills":"sum","deaths":"sum","assists":"sum"}).iloc[1]
    player_assists = player_data.agg({"kills":"sum","deaths":"sum","assists":"sum"}).iloc[2]

    # Partidas do jogador
    player_matchs = player_data.shape[0]

    # Calculo de KDA Total
    player_kda = ((player_kills+player_deaths+player_assists)/3).round(2)

    # Calculo de KDA Medio
    player_mean_kda = (player_kda/player_matchs).round(2)

    # Acrescentando os dados na lista
    playername.append(player)
    kda_total.append(player_kda)
    kda_medio.append(player_mean_kda)
    partidas.append(player_matchs)

# Colocando dentro do df
kda_comparision["playername"] = playername
kda_comparision["quant_partidas"] = partidas 
kda_comparision["kda_total"] = kda_total
kda_comparision["kda_medio"] = kda_medio

# Resultado final
kda_comparision.sort_values(by="kda_medio",ascending=False).head(6)
```
