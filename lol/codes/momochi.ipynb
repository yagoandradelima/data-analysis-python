### Dados individuais Momochi

Abaixo segue o conjunto de informações individuais do jogador do T2 Momochi.

```python
# Criando dataframe de dados individuais do momochi

momochi_data = sup_data[sup_data["playername"] == "Momochi"]
```

```python
# Bloco único para criação do dataframe de KDA dos suportes do Circuito Desafiante

# Criação do df
kda_comparision = pd.DataFrame(columns=["playername","kda_total"])

# Criação das listas que receberão os dados do loop for
playername = []
kda_total = []
kda_medio = []
partidas = []

# loop for que vai percorrer cada nome de jogador e calcular seu KDA
for player in sup_data["playername"].unique():

    player_data = sup_data[sup_data["playername"] == player].copy()

    # Ajustando para não haver 0 na hora de fazer o KDA linha a linha (se houver 0 o resultado pode virar uma divisão com resultado infinito)
    adjusted_deaths = player_data["deaths"].clip(lower=1)

    # Fazendo o kda linha a linha
    player_data["kda"] = (player_data["kills"]+player_data["assists"])/adjusted_deaths



    # Criação do agregador de K, D e A dentro do campeonado
    player_kills = player_data.agg({"kills":"sum","deaths":"sum","assists":"sum"}).iloc[0]
    player_deaths = player_data.agg({"kills":"sum","deaths":"sum","assists":"sum"}).iloc[1]
    player_assists = player_data.agg({"kills":"sum","deaths":"sum","assists":"sum"}).iloc[2]

    # Partidas do jogador
    player_matchs = player_data.shape[0]

    # Calculo de KDA Total
    player_kda = ((player_kills+player_assists)/player_deaths).round(2)

    # Calculo de KDA Medio por partida
    player_mean_kda = player_data["kda"].mean().round(2)

    # Acrescentando os dados na lista
    playername.append(player)
    kda_total.append(player_kda)
    kda_medio.append(player_mean_kda)
    partidas.append(player_matchs)

# Colocando dentro do df
kda_comparision["playername"] = playername
kda_comparision["quant_partidas"] = partidas 
kda_comparision["kda_total"] = kda_total
kda_comparision["kda_medio"] = kda_medio

# Resultado final
kda_comparision.sort_values(by="kda_total",ascending=False).head(10)
```

```python
# KDA Momochi

# Comecei por esse primeira linha e utilizei ela para montar o agregador mais abaixo
momochi_data.agg({"kills":"sum","deaths":"sum","assists":"sum"})

momochi_kills = momochi_data.agg({"kills":"sum","deaths":"sum","assists":"sum"}).iloc[0]
momochi_deaths = momochi_data.agg({"kills":"sum","deaths":"sum","assists":"sum"}).iloc[1]
momochi_assists = momochi_data.agg({"kills":"sum","deaths":"sum","assists":"sum"}).iloc[2]

momochi_kda = round((momochi_kills+momochi_assists)/momochi_deaths, 2)
```

```python
# Quantidade de partidas do jogador no CD
partidas_momochi = momochi_data["league"].value_counts().iloc[0]

# Quantidade diferente de campões escolhidos pelo jogador no CD
quantidade_pool = momochi_data["champion"].value_counts().shape[0]

# Top 3 campeões mais jogados (nomes e partidas)
top_campeoes = momochi_data["champion"].value_counts().nlargest(3).index
top_campeoes_partidas = momochi_data["champion"].value_counts().nlargest(3).values

# kda_médio por partida do Momochi
kda_medio_momochi = kda_comparision[kda_comparision["playername"] == "Momochi"][["kda_medio"]].values[0][0]

# Filtrando os dados dos principais campeões utilizados pelo jogador
momochi_data_top_champs = momochi_data[momochi_data["champion"].isin(top_campeoes)]
```

```python
print("Dados Momochi")
print("Quantidade de partidas pelo Circuito Desafiante em 2025: ", partidas_momochi)
print("Seu KDA Total durante a competição foi de: ", momochi_kda)
print("Seu KDA por partida é de: ", kda_medio_momochi)
print("Total de campeões jogados no Circuito Desafiante: ", quantidade_pool)
print(f"Seus campeões mais jogados foram: {top_campeoes[0]}, {top_campeoes[1]} e {top_campeoes[2]}")
print(f"A quantidade de partidas desses campeões foi: {top_campeoes_partidas[0]}, {top_campeoes_partidas[1]} e {top_campeoes_partidas[2]}")
print("-"*30)
print("WinRate dos campeões mais jogados")
for campeao in top_campeoes:
    vitorias = momochi_data_top_champs[["champion","result"]][(momochi_data_top_champs["champion"] == campeao) & (momochi_data_top_champs["result"] == 1)].shape[0]
    total_partidas = momochi_data_top_champs[["champion","result"]][momochi_data_top_champs["champion"] == campeao].shape[0]

    win_rate = round((vitorias/total_partidas)*100, 2)

    print(f"{campeao} / {win_rate}%")
print("-"*30)
```

### Criação da função geral

```python
# /help dos parâmetros abaixo
# dataframe - O nome do dataframe a ser utilizado para a análise, ele precisa ser derivado do dataframe criado no início desse código e possuir as mesmas colunas

def sup_players_stats(dataframe):

    sup_players = [x for x in dataframe["playername"].unique()]

    # Loopo para pegar a estatistica de cada player
    for player in sup_players:

        player_data = dataframe[dataframe["playername"] == player].copy()

        # ----------------------------------------------------------
        # Montando a lógica para o KDA por partida
        #
        #
        # Primeiro ajustando para que não haja 0 nos cálculos já que na divisão linha a linha ele gera divisões com resultado infinito
        adjusted_deaths = player_data["deaths"].clip(lower=1)

        # Fazendo a coluna calculando o valor linha a linha (a coluna será utilizada na variável kda_partida)
        player_data["kda"] = (player_data["kills"]+player_data["assists"])/adjusted_deaths
        #
        # -----------------------------------------------------------

        # -----------------------------------------------------------
        # Resultados primários
        qtd_partidas = player_data.shape[0]
        qtd_pool = player_data["champion"].unique().shape[0]
        top_picks = [x for x in player_data["champion"].value_counts().nlargest(3).index]
        qtd_match_top_picks = player_data["champion"].value_counts().nlargest(3).values
        kda_mean = round(player_data["kda"].mean(),2)
        #------------------------------------------------------------

        # Print para estilizar o retorno da função (sim, ela não usa o Return)
        print(
            f"""
                Jogador: {player}\n
                Quantidade de partidas: {qtd_partidas}\n
                Quantidade de pool: {qtd_pool}\n
                Campeões mais usados: {top_picks}\n
                Quantidade de partidas jogadas com os campeões: {qtd_match_top_picks}\n
                KDA médio por partida: {kda_mean}
            """
        )

        # Criando o dataframe apenas com os dados dos principais campeões
        player_data_top_champs = player_data[player_data["champion"].isin(top_picks)]

        # Loop for para que existam os dados de WinRate com cada um dos top picks
        for campeao in top_picks:
            vitorias = player_data_top_champs[["champion","result"]][(player_data_top_champs["champion"] == campeao) & (player_data_top_champs["result"] == 1)].shape[0]
            total_partidas = player_data_top_champs[["champion"]][player_data_top_champs["champion"] == campeao].shape[0]
            win_rate = round((vitorias/total_partidas)*100, 2)

            print(
                f"""
                    Campeão: {campeao}\n
                    WinRate: {win_rate}%
                """
                )
```
