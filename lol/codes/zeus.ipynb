```python
import pandas as pd

df_2024 = pd.read_csv("../data/2024_lol_esports_data.csv")

df_2024.head(5)
```

```python
# Verificando os dados para informar o filtro

df_2024["league"].unique()
```

```python
# Filtrando apenas as partidas da LCK que tem dados completos
df_lck = df_2024[(df_2024["league"] == 'LCK') & (df_2024["datacompleteness"] == 'complete')]

df_lck
```

```python
# São inúmeras colunas, mas boa parte delas são estatísticas de equipe
# Preciso selecionar as colunas que são diretamente ligadas à performance de um jogador

df_lck.columns.to_list()
```

```python
colunas_desempenho_por_tempo = ['playername', 'position', 'goldat10', 'xpat10', 'csat10', 'golddiffat10', 'xpdiffat10', 'csdiffat10', 'killsat10', 'assistsat10', 'deathsat10', 'goldat15', 'xpat15', 'csat15','golddiffat15', 'xpdiffat15', 'csdiffat15', 'killsat15',  'assistsat15',  'deathsat15','goldat20', 'xpat20', 'csat20', 'golddiffat20','xpdiffat20', 'csdiffat20', 'killsat20', 'assistsat20', 'deathsat20','goldat25', 'xpat25', 'csat25','golddiffat25', 'xpdiffat25', 'csdiffat25', 'killsat25', 'assistsat25', 'deathsat25']

colunas_info_geral = ['playerid', 'playername', 'position', 'champion', 'damagetochampions', 'dpm', 'damageshare', 'damagetakenperminute', 'damagemitigatedperminute', 'wardsplaced', 'wpm', 'wardskilled', 'wcpm', 'controlwardsbought', 'visionscore', 'vspm', 'totalgold', 'earnedgold','earned gpm', 'total cs', 'cspm', 'kills', 'deaths', 'assists']
```

```python
"""
column_type não necessariamente precisa existir. Tento em vista que o nome das colunas dentro do .agg()
fazem referência apenas àquela lista de colunas. Logo, não necessariamente existe escolha para se colocar
a outra lista de colunas como argumento da função.

Porém mantive pois pretendo adaptar melhor essa função e utilizar vários tipos de colunas diferentes no futuro sem precisar alterar tudo manualmente.

"""

# Criei uma função apenas para organizar o etl em uma única célula

def etl(dataframe, position:str, column_type:list):
    df_position = dataframe[column_type]
    df_position = df_position[df_position["position"] == position]

    df_position = df_position.groupby(["playername"]).agg({"kills":"sum", 
                                                           "assists":"sum",
                                                           "deaths":"sum", 
                                                           "damagetochampions":"mean",
                                                           "dpm":"mean",
                                                           "damagetakenperminute": "mean",
                                                           "damagemitigatedperminute":"mean",
                                                           "visionscore":"mean",
                                                           "totalgold":"mean",
                                                           "total cs":"mean",
                                                           "cspm":"mean"
                                                           })
    
    df_position["kda"] = (df_position["kills"] + df_position["assists"]) / df_position["deaths"]

    nova_ordem = ['kda', 'kills', 'deaths', 'assists',  'damagetochampions', 'dpm', 'damagetakenperminute', 'damagemitigatedperminute', 'visionscore', 'totalgold', 'total cs','cspm']

    df_position = df_position[nova_ordem]

    return df_position.round(2).reset_index()
```

```python
# Particularmente uma organização odiável, mas útil para ler os processos sem precisa rolar a tela

df_top_players = etl(df_lck, 'top', colunas_info_geral)
```

```python
df_top_players.sort_values(by=["kda", 
                                "kills", 
                                "deaths", 
                                "assists",
                                "damagetochampions"], 
                                ascending=[False, 
                                            False, 
                                            False,
                                            False, 
                                            False
                                        ]).reset_index(drop=True, inplace=True)
```

```python
df_top_players
```

```python
df_top_players.to_csv("../data/top_players.csv")
```

```python
def etl_t(dataframe, position:str, column_type:list):
    df_position = dataframe[column_type]
    df_position = df_position[df_position["position"] == position]


    df_position = df_position.groupby(["playername"]).agg({item:"mean" for item in colunas_desempenho_por_tempo if df_lck[item].dtype == 'float64'})

    return df_position.round(2).reset_index()    
```

```python
dft_top_players = etl_t(df_lck, 'top', colunas_desempenho_por_tempo)
```
